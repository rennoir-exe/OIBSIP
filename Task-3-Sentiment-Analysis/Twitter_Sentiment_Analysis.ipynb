```python
import pandas as pd

df = pd.read_csv("Twitter_Data.csv")
df.head()
```

```python
df.info()
```

```python
df['category'].value_counts()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

df['category'].value_counts().plot(kind='bar')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment Category')
plt.ylabel('Number of Tweets')
plt.show()
```

```python
df['Sentiment'] = df['category'].map({
    1.0: 'Positive',
    0.0: 'Neutral',
    -1.0: 'Negative'
})

df[['clean_text', 'Sentiment']].head()
```

### Conclusion

In this project, Twitter data was analyzed to understand public sentiment.
The dataset was cleaned by removing missing values, and the distribution
of sentiments was explored using visualizations.

The results show that tweets can be grouped into positive, neutral,
and negative categories. This type of sentiment analysis can help
